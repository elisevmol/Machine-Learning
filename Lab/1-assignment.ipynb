## Problem 3: Implement Linear regression

In class you looked at performing regression using gradient descent. Now you are going to implement it.

```python
# Run this to get all the variables 
x_test = [[3],[5],[6]]
y_test = [[6],[7],[10]]
```

```python
#Make a prediction function h
def prediction_function(x, theta0, theta1):
    return theta0 + theta1 * x
```

```python
#Use the output of that function to compute the cost function J:
def cost_function(y, theta0, theta1):
    # x_predict is an array of all the values that are outputed by prediction_function(x, theta0, theta1) 
    # this looks like [3, 5, 6]
    # y is y1 or y2 or another array of y values
    # this looks like [[6], [7], [10]]
    # output looks like [ 32.52232005]
    sumpart = 0
    length = len(y)
    one_over_2m = float(1) / (float(2) * float(length))
    while length > 0: 
        sumpart += square(prediction_function(x[length - 1], theta0, theta1) - y[length - 1])
        length -= 1
    return one_over_2m * sumpart

print cost_function(y_test, 0, 1)
```

```python
#Create a function that returns the gradient values, given h (x_predict), y and x:
def compute_gradient(x, y, theta0, theta1):
    # x_predict is an array of all the values that are outputed by prediction_function(x, theta0, theta1) 
    # this looks like [3, 5, 6]
    # y is y1 or y2 or another array of y values
    # this looks like [[6], [7], [10]]
    # x is x1 or x2 or another array of x values
    # this looks like [[6],[7],[10]]
    # output is gradient for theta0 and theta1, looks like (-43, -9)
    gradient0 = 0
    gradient1 = 0 
    length = len(x)
    while length > 0:
        gradient0 += (prediction_function(x[length - 1][0], theta0, theta1) - y[length - 1])
        gradient1 += (prediction_function(x[length - 1][0], theta0, theta1) - y[length - 1]) * prediction_function(x[length - 1][0], theta0, theta1)
        length -= 1
    return gradient0, gradient1

print compute_gradient(x1, y1, 0, 1)
```

```python
def total_gradient_descent(alpha, theta0, theta1, x, y, iterations, stopcondition):
    length = float(len(x))
    one_over_m = float(1) / (float(2) * length)
    i = 0
    cost_function_outcomes = []
    while (i < iterations):
        zero = alpha * one_over_m * compute_gradient(x, y, theta0, theta1)[0]
        one = alpha * one_over_m * compute_gradient(x, y, theta0, theta1)[1]
        theta0 = theta0 - zero
        theta1 = theta1 - one
        cost_function_outcomes.extend([[cost_function(y, theta0, theta1)[0], i]])
        i += 1
    x_axis = [i[1] for i in cost_function_outcomes]
    y_axis = [i[0] for i in cost_function_outcomes]
    plt.plot(x_axis, y_axis, 'o')
    plt.show
    return theta0, theta1

#default values
alpha = 0.1
theta0 = 0
theta1 = 1
x = x1
y = y1
iterations = 300
stopcondition = 1

#test with default values
total_gradient_descent(alpha, theta0, theta1, x, y, iterations, stopcondition)
```
