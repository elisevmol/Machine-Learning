## Problem 3: Implement Linear regression

In class you looked at performing regression using gradient descent. Now you are going to implement it.

Make sure to comment your code!

```python
def prediction_function(x, theta0, theta1):
    """takes x wich is of type 'numpy.ndarray' which looks like [-2.71575179  0.61715501]
    and theta0, theta1 which are numbers"""
    return theta0 + theta1 * x
```

```python
def cost_function(x_predict,y):
    """x_predict is of the type numpy.ndarray and looks like array([ 10.51861267,  10.17518207 ])
    and y is of type 'numpy.ndarray' which looks like [-203.93310997   44.01211128]
    output is of type numpy.float64, which is a number"""
    m = len(y)
    summedvalues = 0
    prefix = 1 / float(2 * m)
    while m > 0:
        summedvalues += square(x_predict[m - 1] - y[m - 1])
        m -= 1
    return prefix * summedvalues
```

```python
def compute_gradient(x_predict,y, x):
    """ y is of type 'numpy.ndarray' which looks like [-203.93310997   44.01211128]
        x is of type 'numpy.ndarray' which looks like [-2.71575179  0.61715501]
        x_predict is of type numpy.ndarray and looks like array([ 10.51861267,  10.17518207 ])
        output is of type tuple and looks like (207.82887001865009, -37.805127139691692)"""
    gradient0 = 0
    gradient1 = 1
    m = len(x)
    while m > 0:
        gradient0 += (x_predict[m - 1] - y[m - 1])
        gradient1 += (x_predict[m - 1] - y[m - 1]) * x_predict[m - 1]
        m -= 1
    return gradient0, gradient1
```

```python
#The default parameters
alpha = 0.01
theta0 = 1
theta1 = 1
x = x5.flatten()
y = y5
iterations = 50
stopcondition = 0.1

#The values I have defined to use in the final code
prefix = 1 / float(len(x) * 2)
output_cost_function = []
i = 0
cost = 10

while (i < iterations) and (cost > stopcondition):
    x_predict = prediction_function(x, theta0, theta1) #setting x_predict again with the updated theta's
    i += 1 #increasing the amount of iterations
    zero = alpha * prefix * compute_gradient(x_predict, y, x)[0] #calculating the part that needs to be subtracted from theta0
    one = alpha * prefix * compute_gradient(x_predict, y, x)[1] #calculation the part that needs to be subtracted from thata1
    theta0 = theta0 - zero #updating theta0
    theta1 = theta1 - one #updating theta1
    cost = cost_function(x_predict, y) #updating cost
    output_cost_function.append(cost) #adding the output of the cost function to the list so we can plot the values after
    

#plotting the output of the cost function at each iteration
plt.plot(output_cost_function, 'o')
plt.show()
    
#plotting the prediction vs the data
plt.plot(x, prediction_function(x,theta0,theta1))
plt.plot(x, y, 'o')
plt.show()
```
